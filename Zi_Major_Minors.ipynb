```python
num_minor =[]
for ind in demo.index:
    minor = 0;
    if not(pd.isnull(demo.loc[ind, 'Degree Minor1 Code']) or demo['Degree Minor1 Code'][ind]=="."):
        minor = minor +1
    if not (pd.isnull(demo.loc[ind, 'Degree Minor2 Code']) or demo['Degree Minor2 Code'][ind]=="."):
        minor = minor +1
    if not (pd.isnull(demo.loc[ind,'Degree Minor3 Code']) or demo['Degree Minor3 Code'][ind]=="."):
        minor = minor +1
    
    num_minor.append(minor)

demo['Num of Minors'] = num_minor
```

```python
#GETTING THE NUMBER OF MAJOR OF EACH ROW
```

```python
num_major =[]
for ind in demo.index:
    major = 0;
    if not(pd.isnull(demo.loc[ind, 'Degree Major1 Code']) or demo['Degree Major1 Code'][ind]=="."):
        major = major +1
    if not (pd.isnull(demo.loc[ind, 'Degree Major2 Code']) or demo['Degree Major2 Code'][ind]=="."):
        major = major +1

    
    num_major.append(major)

demo['Num of Majors'] = num_major
```
